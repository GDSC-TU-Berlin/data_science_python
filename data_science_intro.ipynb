#### 1) Happiness Classification dataset

```python
# Task 1) Load your data and inspect the data via displaying the first row and find out what understand the scale of the dataset

# solution: 
data_happy = pd.read_csv('happydata.csv')
data_happy.head()
# likert scale 0-5, rating of a the city where they live in regard of happiness, happiness is rated binary
```

```python
# Before diving into the analysis, it's always a good idea to explore the basic properties of your dataset.
# Task 2) basic statistics, what are the results of the functionality .describe()

# solution: 
data_happy.describe()
```

```python
# Visualisation of data
# Task 3) visualize the distribution of happiness
data_happy['happy'].value_counts().plot(kind='bar', color=['salmon', 'lightblue'])
plt.title('Distribution of Happiness')
plt.xticks(ticks=[0,1], labels=['Unhappy', 'Happy'], rotation=0)
plt.ylabel('Count')
plt.show()
```

```python
# Relation between the different categories: plotting the average rating for each catebory based on happiness

metrics = ['housecost', 'schoolquality', 'policetrust', 'streetquality']

for metric in metrics:
    data_happy.groupby('happy')[metric].mean().plot(kind='bar', color=['salmon', 'lightblue'])
    plt.title(f'Average rating for {metric}')
    plt.xticks(ticks=[0,1], labels=['Unhappy', 'Happy'], rotation=0)
    plt.ylabel('Average Rating')
    plt.show()
```
